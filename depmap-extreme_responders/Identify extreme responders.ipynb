```r
blood_lines <-
    c(
        'thyroid',
        'haematopoietic_and_lymphoid_tissue',
        'Thymus',
        'ventricle',
        'cord blood',
        'B lymphoblastoid cells',
        'lymph_node'
        
    )
```

```r
solid_lines <-
    model_tbl %>%
    filter(
        !is.na(`SampleCollectionSite`),
        !(`SampleCollectionSite` %in% blood_lines)
    )

solid_lines
```

```r
system.time({

    cnv_wtbl <-
        read_csv(
            'data/OmicsCNGeneWGS.csv',
            col_types=cols(
                `...1`=col_integer(),
                `SequencingID`=col_character(),
                `ModelID`=col_character(),
                `IsDefaultEntryForModel`=col_factor(levels=c('Yes', 'No')),
                `ModelConditionID`=col_character(),
                `IsDefaultEntryForMC`=col_factor(levels=c('Yes', 'No')),
                .default=col_double()
            )
        ) %>%
        select(-`...1`)
    
})

cnv_wtbl %>%
    head(20)
```

```r
system.time({

    depmap_wtbl <-
        read_csv(
            'data/CRISPRGeneEffect.csv',
            col_types=cols(`...1`=col_character(), .default=col_double())
        ) %>%
        rename(`ModelID`=`...1`) %>%
        filter(`ModelID` %in% solid_lines$`ModelID`)
    
})

depmap_wtbl %>%
    head(20)
```

```r
depmap_wtbl %>%
    nrow()
```

```r
heme_genes <-
    bind_rows(
        read_csv('lymph-context-impact.csv'),
        read_csv('myleoid-context-impact.csv')
    ) %>%
    distinct(`entity`) %>%
    pull()

heme_genes
```

```r
gene_means <-
    depmap_wtbl %>%
    select(-`ModelID`) %>%
    colMeans(na.rm=TRUE) %>%
    (function(x) {
        tibble(
            `gene`=names(x),
            `mean_ceres`=as.vector(x)
        ) %>%
        arrange(`mean_ceres`)
    })()

gene_means %>%
    head()
```

```r
gene_deplines <-
    depmap_wtbl %>%
    select(-`ModelID`) %>%
    (function(x) {
        (as.matrix(x) < -1.5) %>%
            colSums(na.rm=TRUE)
    }) %>%
    (function(x) {
        tibble(
            `gene`=names(x),
            `no_of_deplines`=as.vector(x)
        ) %>%
        arrange(`no_of_deplines`)
    })()

gene_deplines %>%
    tail(50)
```

```r
gene_obs <-
    depmap_wtbl %>%
    select(-`ModelID`) %>%
    (function(x) {
        as.matrix(x) %>%
            is.na() %>%
            (function(x) { !x }) %>%
            colSums(na.rm=TRUE)
    }) %>%
    (function(x) {
        tibble(
            `gene`=names(x),
            `no_of_obs`=as.vector(x)
        ) %>%
        arrange(`no_of_obs`)
    })()

gene_obs
```

```r
goi_tbl <-
    gene_means %>%
    left_join(gene_deplines, by='gene') %>%
    left_join(gene_obs, by='gene') %>%
    filter(
        `mean_ceres` > -0.3,
        `mean_ceres` < 0.3,
        `no_of_deplines` > 0,
        `no_of_deplines` < 4,
        `no_of_obs` > 1000,
        !(`gene` %in% heme_genes)
    )

goi_tbl
```

```r
depmap_goi_tbl <-
    depmap_wtbl %>%
    select(`ModelID`, all_of(goi_tbl$`gene`)) %>%
    pivot_longer(
        names_to='gene',
        values_to='dep',
        -`ModelID`
    ) %>%
    arrange(`gene`, `dep`) %>%
    left_join(goi_tbl, by='gene') %>%
    left_join(model_tbl %>% select(`ModelID`, `StrippedCellLineName`), by='ModelID') %>%
    relocate(`StrippedCellLineName`)

depmap_goi_tbl
```

```r
n1s_tbl <-
    depmap_goi_tbl %>%
    group_by(`gene`) %>%
    slice_head(n=1) %>%
    ungroup() %>%
    add_count(`ModelID`) %>%
    filter(`n` == 1, `no_of_deplines` == 1) %>%
    arrange(`dep`)

n1s_tbl
```

```r
cnv_goi_tbl <-
    cnv_wtbl %>%
    filter(`ModelID` %in% n1s_tbl$`ModelID`) %>%
    select(`ModelID`, any_of(n1s_tbl$`gene`)) %>%
    pivot_longer(
        names_to='gene',
        values_to='cnv',
        -`ModelID`
    ) %>%
    drop_na()

cnv_goi_tbl
```

```r
ns1_tbl_with_cnv <-
    n1s_tbl %>%
    left_join(cnv_goi_tbl, by=c('ModelID', 'gene')) %>%
    group_by(`StrippedCellLineName`) %>%
    slice_max(`cnv`, n=1) %>%
    ungroup() %>%
    arrange(`dep`)

ns1_tbl_with_cnv
```

```r
ns1_tbl_with_cnv %>%
    filter(`cnv` > 3)
```

```r
ns1_tbl_with_cnv %>%
    mutate(`gene`=str_replace_all(`gene`, ' \\(.*\\)', '')) %>%
    write_tsv('results/n1s_tbl.tsv')
```
