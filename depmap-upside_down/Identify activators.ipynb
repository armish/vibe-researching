```r
cancerous_lines <-
    model_tbl %>%
    filter(
        is.na(`EngineeredModel`),
        `ModelType` != 'Organoid',
        !(`OncotreePrimaryDisease` %in% c(
          'Mature B-Cell Neoplasms', 'Acute Myeloid Leukemia', 'B-Cell Acute Lymphoblastic Leukemia',
          'T-Lymphoblastic Leukemia/Lymphoma', 'Myeloproliferative Neoplasms', 'Mature T and NK Neoplasms',
          'Non-Hodgkin Lymphoma', 'Hodgkin Lymphoma',
          'Non-Cancerous'
        ))
    )

cancerous_lines
```

```r
system.time({

    depmap_wtbl <-
        read_csv(
            'data/CRISPRGeneEffect.csv',
            col_types=cols(`...1`=col_character(), .default=col_double())
        ) %>%
        rename(`ModelID`=`...1`)
    
})

depmap_wtbl %>%
    head(20)
```

```r
non_cancerous_lines %>%
    filter(`ModelID` %in% depmap_wtbl$`ModelID`) %>%
    t()
```

```r
depmap_wtbl %>%
    nrow()
```

```r
gene_deplines <-
    depmap_wtbl %>%
    filter(`ModelID` %in% cancerous_lines$`ModelID`) %>%
    select(-`ModelID`) %>%
    (function(x) {
        (as.matrix(x) < -1.5) %>%
            colSums(na.rm=TRUE)
    }) %>%
    (function(x) {
        tibble(
            `gene`=names(x),
            `no_of_deplines`=as.vector(x)
        ) %>%
        arrange(`no_of_deplines`)
    })()

gene_deplines %>%
    tail(50)
```

```r
gene_obs <-
    depmap_wtbl %>%
    filter(`ModelID` %in% cancerous_lines$`ModelID`) %>%
    select(-`ModelID`) %>%
    (function(x) {
        as.matrix(x) %>%
            is.na() %>%
            (function(x) { !x }) %>%
            colSums(na.rm=TRUE)
    }) %>%
    (function(x) {
        tibble(
            `gene`=names(x),
            `no_of_obs`=as.vector(x)
        ) %>%
        arrange(`no_of_obs`)
    })()

gene_obs
```

```r
depmap_wtbl %>%
    filter(`ModelID` %in% non_cancerous_lines$`ModelID`)
```

```r
depmap_wtbl %>%
    filter(
        `ModelID` %in% non_cancerous_lines$`ModelID`
    ) %>%
    select(`ModelID`, `APC (324)`, `RBM5 (10181)`, `KCTD5 (54442)`)
```

```r
nc_gene_means <-
    depmap_wtbl %>%
    filter(`ModelID` %in% non_cancerous_lines$`ModelID`) %>%
    select(-`ModelID`) %>%
    colMeans(na.rm=TRUE) %>%
    (function(x) {
        tibble(
            `gene`=names(x),
            `mean_ceres`=as.vector(x)
        ) %>%
        arrange(`mean_ceres`)
    })() %>%
    rename(`nc_mean_ceres`=`mean_ceres`)

nc_gene_means %>%
    head()
```

```r
c_gene_means <-
    depmap_wtbl %>%
    filter(`ModelID` %in% cancerous_lines$`ModelID`) %>%
    select(-`ModelID`) %>%
    colMeans(na.rm=TRUE) %>%
    (function(x) {
        tibble(
            `gene`=names(x),
            `mean_ceres`=as.vector(x)
        ) %>%
        arrange(`mean_ceres`)
    })() %>%
    rename(`c_mean_ceres`=`mean_ceres`)

c_gene_means %>%
    head()
```

```r
heme_and_myleoid_genes <-
    bind_rows(
        read_csv('lymph-context-impact.csv'),
        read_csv('myleoid-context-impact.csv')
    ) %>%
    pull(`entity`) %>%
    unique()

heme_and_myleoid_genes
```

```r
goi_tbl <-
    left_join(nc_gene_means, c_gene_means, by='gene') %>%
    left_join(gene_obs, by='gene') %>%
    left_join(gene_deplines, by='gene') %>%
    filter(
        `nc_mean_ceres` > 0.2,
        `c_mean_ceres` > -0.5,
        `no_of_deplines` > 0,
        `no_of_obs` > 1000
    ) %>%
    arrange(-`nc_mean_ceres`) %>%
    mutate(`heme_or_myeloid_bias`=`gene` %in% heme_and_myleoid_genes)

goi_tbl
```

```r
goi_tbl %>%
    filter(!`heme_or_myeloid_bias`)
```
